```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

df = pd.read_csv('./public_data/subwaytime.csv', header=[0,1] )
df
```

```python
commute_time_df = df.iloc[:,[1,3,35,37]]

ctd = commute_time_df.copy()
ctd

ctd.columns = ['호선명', '지하철명', '19-20시 하차', '20-21시 하차']
ctd

ctd['19-21시 하차'] = ctd['19-20시 하차'] + ctd['20-21시 하차']
ctd
```

```python
grouped_ctd = ctd.groupby('지하철명')['19-21시 하차'].sum().reset_index()
grouped_ctd
```

```python
top5_stations = grouped_ctd.sort_values('19-21시 하차', ascending=False).head(5)

top5_stations
```
